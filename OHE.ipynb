```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/cars.csv")
df.sample(5)
```

```python
pd.get_dummies(df ,columns = ['fuel','owner'] ,drop_first= True ,) #temporary not in place
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
ohe = OneHotEncoder(drop='first')
encoded_array = ohe.fit_transform(df[['fuel', 'owner']]).toarray()
encoded_df = pd.DataFrame(encoded_array, columns=ohe.get_feature_names_out(['fuel', 'owner']))
final_df = pd.concat([df[['brand', 'km_driven']].reset_index(drop=True), encoded_df], axis=1)
final_df
```

```python
counts = final_df['brand'].value_counts()
threshold = 100
repl = counts[counts<=threshold].index
lis = pd.get_dummies(final_df['brand'].replace(repl,'uncommon'))
pd.concat([lis, df['km_driven'].reset_index(drop=True), encoded_df], axis=1)
```
